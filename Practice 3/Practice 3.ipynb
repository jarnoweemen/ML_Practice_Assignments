```python
# Necessary
import pandas as pd

# Natural language processing
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Sklearn
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df_twitter = pd.read_csv("Data/MeTooHate.csv")
df_twitter.head()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
tokens = [tokenizer.tokenize(str(text)) for text in df_twitter["text"]]
tokens
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words("english"))
filtered_words = [[w for w in t if not w.lower() in stop_words] for t in tokens]
filtered_words
```

```python
ps = PorterStemmer()
stemmed_words = [[ps.stem(w) for w in fw] for fw in filtered_words]
stemmed_words
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
wnl = WordNetLemmatizer()
lemma_words = [[wnl.lemmatize(w) for w in fw] for fw in filtered_words]
lemma_words
```

```python
lemma_to_lower = [[w.lower() for w in lw] for lw in lemma_words]
lemma_to_lower
```

```python
df_twitter["text"] = [' '.join(ltl) for ltl in lemma_to_lower]
df_twitter.head()
```

```python
df_twitter["location"].unique()
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_twitter["text"])

model = NMF(n_components=10, random_state=42)
model.fit(tfidf_matrix)
```

```python
feature_names = vectorizer.get_feature_names()
topics = []
for topic_idx, topic in enumerate(model.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-6:-1]]  # Get the top 5 words for each topic
    topics.append(top_words)

# Print the topics and associated words
for topic_idx, topic in enumerate(topics):
    print(f"Topic {topic_idx+1}: {' '.join(topic)}")
```
